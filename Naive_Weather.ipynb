Weather prediction using Naive Bayes (Gaussian)                                             
Bayes Theorem: P(A|B)=P(B|A)* P(A)/P(B)


```python
#Required Modules
import pandas as pd
from sklearn.preprocessing import LabelEncoder                   #In ML algo are not trained to reading the strings therefore LabelEncoder converts 
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("new_dataset.csv")                 #df = dataframe
df
```

```python
#Encoding the strings to Numericals
Numerics = LabelEncoder()
```

```python
#Dropping the target variable and mae it is as newfram
inputs=df.drop('Play',axis='columns')
target=df['Play']
target
```

```python
inputs['Outlook_n']=Numerics.fit_transform(inputs['Outlook'])
inputs['Temp_n']=Numerics.fit_transform(inputs['Temp'])
inputs['Hum_n']=Numerics.fit_transform(inputs['Humidity'])
inputs['Wind_n']=Numerics.fit_transform(inputs['Windy'])
inputs
```

```python
#Dropping the string values
inputs_n=inputs.drop(['Outlook','Temp','Humidity','Windy'],axis='columns')
inputs_n
```

```python
#Applying the Gaussian Naivebayes
Classifier=GaussianNB()
Classifier.fit(inputs_n,target)
```

```python
Classifier.score(inputs_n,target)        #85% Accuracy
```

```python
#Prediction
Classifier.predict([[0,0,0,1]])
```
