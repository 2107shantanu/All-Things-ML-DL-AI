```python
#Naive Bayes Titanic Data Set
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
df=pd.read_csv("train.csv")
df.head()                                       #Calling header part not whole df
```

```python
#Dropping Unuseful Info
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)  
#When inplace = True ,the data is modified in place, which means it will return nothing and the dataframe is now updated.
# When inplace = False , which is the default, then the operation is performed and it returns a copy of the object.
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
target.head()
```

```python
Numerics = LabelEncoder()
inputs['Sex_n']=Numerics.fit_transform(inputs['Sex'])
inputs
```

```python
inputs_n=inputs.drop(['Sex'],axis='columns')
inputs_n
```

```python
inputs_n.columns[inputs_n.isna().any()]   #Checking for any NA value in df = inputs_n
```

```python
inputs_n.Age = inputs_n.Age.fillna(inputs_n.Age.mean())       #fill in na values with mean ages
inputs_n.head()
```

```python
from sklearn.model_selection import train_test_split     #Spiliting Training and Testing data in 30-70 ratio
X_train, X_test, y_train, y_test = train_test_split(inputs_n,target,test_size=0.3)   #Focus on inputs_n and target
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
